## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Data augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_d=ImageDataGenerator(rescale=(1./255),horizontal_flip=True,shear_range=0.4)
test_d=ImageDataGenerator(rescale=(1./255),horizontal_flip=True,shear_range=0.4)
```

```python
train=train_d.flow_from_directory('/kaggle/input/bird-species-classification/train_data/train_data',target_size=(100,100),class_mode='categorical',batch_size=5)
test=test_d.flow_from_directory('/kaggle/input/bird-species-classification/test_data/test_data',target_size=(100,100),class_mode='categorical',batch_size=5)
```

```python
train.class_indices
```

## Building the model -CNN

```python
from tensorflow.keras.layers import Convolution2D,MaxPooling2D,Flatten,Dense
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Convolution2D(20,(3,3),activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(59,activation='relu'))
model.add(Dense(16,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train,batch_size=5,validation_data=test,epochs=30)
```

```python
model.save('birdclass.h5')
     
```

## Model Tuning

```python
model = Sequential()
model.add(Convolution2D(13,(3,3),activation='relu',input_shape=(100, 100, 3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Convolution2D(26,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Convolution2D(39,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(56,activation='relu'))
model.add(Dense(40,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(16,activation='softmax'))
```

```python
model.summary()
     
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train,batch_size=5,validation_data=test,epochs=35)
```

## Pre-trained model -VGG16

```python
from tensorflow.keras.layers import Dense,Flatten,Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
import numpy as np
```

```python
train_d=ImageDataGenerator(rescale=(1./255),horizontal_flip=True,shear_range=0.4,zoom_range=0.2)
test_d=ImageDataGenerator(rescale=(1./255))
```

```python
train=train_d.flow_from_directory('/kaggle/input/bird-species-classification/train_data/train_data',target_size=(224,224),class_mode='categorical',batch_size=25)
test=test_d.flow_from_directory('/kaggle/input/bird-species-classification/test_data/test_data',target_size=(224,224),class_mode='categorical',batch_size=25)
```

```python
from tensorflow.keras.applications.vgg16 import VGG16,preprocess_input
```

```python
vgg= VGG16(include_top=False,input_shape=(224,224,3))
```

```python

for layer in vgg.layers:
    layer.trainable=False 
```

```python
x=Flatten()(vgg.output)
```

```python
prediction=Dense(16,activation='softmax')(x)
```

```python
model=Model(inputs=vgg.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit_generator(train,validation_data=test,epochs=20,steps_per_epoch=len(train),validation_steps=len(test))
```
